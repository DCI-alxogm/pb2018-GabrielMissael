```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as con
import math

%matplotlib inline 
```

```python
puntos = np.loadtxt("IRCF.txt").T
y=puntos[3]
```

```python
for i in range(np.size(y)):
        if((puntos[1][i]-y[i])<0):
            y[i]=puntos[1][i]
```

```python
puntos = np.loadtxt("IRCF.txt").T
plt.errorbar(puntos[0], puntos[1], [y,puntos[3]], fmt='o')
plt.ylabel("Intensidad(MJy/sr)")
plt.xlabel("Frecuencia(1/cm)")
plt.grid()
```

```python
def Int(v, T):
    h = con.h
    c=con.c
    k=con.k
    I = (h*v**3)/((c**2)*(math.exp((h*v)/(k*T))-1))
    I /= (10**(-26))
    return I
```
