```python
#Creado por Gabriel Missael Barco
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
n = 20
#segundos de analizis
h = 0.001
#cambio de tiempo
ini = 20.0
#amplitud
s= n/h
#numero total de pasos
s=int(s)
w = 3
#frecuencia angular

v = np.linspace(0, s, s+1)
#arreglo para las velocidades
the = np.linspace(0, s, s+1)
#arreglos para los angulos
t = np.linspace(0, s, s+1)

v[0] = 0.0
the[0]= ini
#Igualamos la primera velocidad a 0 y el primer angulo a la amplitud
```

```python
def f1(ang):
    return h*(-w**2*ang)
#Esta funcion devuelve una velocidad de acuerdo a un angulo y a la 
#velocidad en un momento anterior

def f2(v):
    return h*v
#Esta funcion devuelve un angulo de <cuerdo a la velocidad y al
#angulo del momento anterior

def armonico():
    #Esta funcion realiza todos lo valores de angulos y velocidades
    for i in range(s):
        v[i+1] = v[i]+f1(the[i])
        #Nuevo valor de la velocidad
        the[i+1]= the[i]+f2(v[i])
        #Nuevo valor del angulo
        if(the[i+1]>ini):
            the[i+1]=ini
        if(the[i+1]<-ini):
            the[i+1]=-ini
        #Estos condicionales evitan que el angulo sobrepase la amplitud 
        #maxima
```

```python
armonico()
```

```python
plt.plot(t/1000, the)
plt.grid()
plt.xlabel("t(s)")
plt.ylabel("theta(°)")
#Graficamos tiempo vs. vangulo
```

```python
plt.plot(t/1000, v)
plt.grid()
plt.xlabel("t(s)")
plt.ylabel("v(°/s)")
#Graficamos tiempo vs. velocidad
```

```python
plt.plot(v, the)
plt.grid()
plt.xlabel("v(°/s)")
plt.ylabel("theta(°)")
#Graficamos velocidad vs. angulo
```
